```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('test.csv')
```

```python
kj=df1.copy()
```

```python
df1.info()
```

```python
plt.figure(figsize=(20, 10))
cmap = sns.cubehelix_palette(light=1, as_cmap=True, reverse=True)
sns.heatmap(df1.isnull(), cmap=cmap)
```

```python
df1.drop(['Id','Alley','PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
```

```python
df1.info()
```

```python
for label,content in df1.items():
    if not pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            #fill missing values with median
            df1[label]=content.fillna(content.mode()[0])
```

```python
for label,content in df1.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            #fill missing values with median
            df1[label]=content.fillna(content.median())
```

```python
df1.info()
```

```python
sk=pd.read_csv('processed_train_data')
```

```python
sk.info()
```

```python
sk.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
final=pd.concat([sk,df1],axis=0)
```

```python
final.info()
```

```python
# selecting object dtypes in the data
qw=final.select_dtypes(include=['object'])
qw
```

```python
we=final.select_dtypes(include=['int64','float64'])
we
```

```python
#get dummies of object data
X = pd.get_dummies(qw, prefix_sep='_', drop_first=True)
X
```

```python
ff=pd.concat([X,we],axis=1)
```

```python
ff.head()
```

```python
ff=ff.loc[:,~ff.columns.duplicated()]
```

```python
ff.info()
```

```python
ty=ff.drop('SalePrice',axis=1)
```

```python
x=ty.iloc[:1460,:]
final_test=ty.iloc[1460:,:]
```

```python
op=ff['SalePrice']
```

```python
y=op.iloc[:1460]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# import random forest regressor
np.random.seed(42)
import xgboost as xgb 
model=xgb.XGBRegressor()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
scores = cross_val_score(model, x, y, cv=5)
scores
```

```python
from sklearn.metrics import mean_squared_log_error
mean_error=np.sqrt(mean_squared_log_error(y_test,pred))
```

```python
mean_error
```

```python
final_predictions=model.predict(final_test)
```

```python
final_predictions
```

```python
df_pred=pd.DataFrame()
df_pred["Id"]=kj["Id"]
df_pred['SalePrice']=final_predictions
df_pred
```

```python
df_pred.to_csv("xgboost_predictions.csv",index=False)
```

#  XGBoost Hyper parameters tunning

```python
params={
        "learning_rate": [0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40],
        "max_depth": [3,4,5,6,8,10,12,15],
        "min_child_weight": [1,3,5,7],
        "n_estimators":[10,30,50,100,150],
        "gamma": [0.0,0.1,0.2,0.3,0.4],
        "colsample_bytree": [0.3,0.4,0.5,0.7],
        "random_state":[42]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search=RandomizedSearchCV(model,param_distributions=params,n_iter=5,n_jobs=-1,cv=5,verbose=3)
random_search.fit(x_train,y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_params_
```

```python
classifier=xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=0.7, gamma=0.3, gpu_id=-1,
             importance_type='gain', interaction_constraints='',
             learning_rate=0.2, max_delta_step=0, max_depth=12,
             min_child_weight=5, monotone_constraints='()',
             n_estimators=150, n_jobs=0, num_parallel_tree=1, random_state=42,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, subsample=1,
             tree_method='exact', validate_parameters=1, verbosity=None)
```

```python
classifier.fit(x_train,y_train)
```

```python
pred=classifier.predict(x_test)
```

```python
error=np.sqrt(mean_squared_log_error(y_test,pred))
```
